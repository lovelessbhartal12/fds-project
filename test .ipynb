```python
from sklearn.utils import resample

# Separate the majority and minority classes
df_majority = df[df['FraudFound_P'] == 0]
df_minority = df[df['FraudFound_P'] == 1]

# Resample the minority class
df_minority_upsampled = resample(df_minority, replace=True, n_samples=24390, random_state=0)

# Combine the majority class with the upsampled minority class
df_resampled = pd.concat([df_majority, df_minority_upsampled])

# Check the new class distribution
print(df_resampled['FraudFound_P'].value_counts())
```

```python
sns.countplot(x=df_resampled['FraudFound_P'].head(10))  # Limit the size for testing

# Show the plot
plt.show()
```

```python
#Removing outliers using z-score
from scipy import stats
import numpy as np

# Assuming df is your DataFrame
# Step 1: Calculate Z-scores for each feature (ignoring non-numeric columns)
z_scores = np.abs(stats.zscore(df_resampled))

# Step 2: Define a threshold for outliers (typically 3)
threshold = 3

# Step 3: Filter out the outliers
df_no_outliers = df_resampled[(z_scores < threshold).all(axis=1)]

# Check the shape of the data before and after removing outliers
print("Original shape:", df.shape)
print("New shape:", df_no_outliers.shape)
```
